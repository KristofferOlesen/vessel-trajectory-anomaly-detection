```python
# Define the current setup
generative_dist = "GMM"
learning_rate = 0.00003
GMM_equally_weighted = False
scheduler_gamma = [0.7, 0.5, 0.6, 0.6]
scheduler_milestones = [600, 800, 1000, 1300]
GMM_components = 4

# Use the SummaryModels class
summary_models = SummaryModels(file_name, generative_dist=generative_dist, learning_rate=learning_rate,
                               GMM_equally_weighted=GMM_equally_weighted, scheduler_gamma=scheduler_gamma,
                               scheduler_milestones=scheduler_milestones, GMM_components=GMM_components)
df_GMM_4 = summary_models.load_curves_df("GMM_4")
summary_models.plot_curves(df_default)
summary_models.plot_curves(df_default, ylims=[(4, 40), (0, 10), (-40, -4)])
df_GMM_4 = summary_models.load_curves_df("GMM_4", level="Step")
```

# Comparison

```python
df_GMM_4 = df_GMM_4[:3160]
df_GMM_3['Epoch'] = df_GMM_4['Number of optimiser steps']
df_GMM_3.columns = df_GMM_4.columns
df = pd.concat([df_GMM_3, df_GMM_4])
df.reset_index(drop=True, inplace=True)
hue = "Setup type"
hue_order = ["GMM_3", "GMM_4"]
x="Number of optimiser steps"
ylims=[(4, 40), (0, 10), (-40, -4)]
xlims=[(0, 45000), (0, 45000), (0, 45000)]
summary_models.plot_curves(df[df["Data set type"] == "Training"], hue, hue_order, x=x,
                           title="Training data set", ylims=ylims, xlims=xlims)
summary_models.plot_curves(df[df["Data set type"] == "Validation"], hue, hue_order, x=x,
                           title="Validation data set", ylims=ylims, xlims=xlims)
```
