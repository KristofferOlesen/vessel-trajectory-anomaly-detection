# Load packages and data

```python
# Loading the required packages to run this notebook
from src.visualisation.SummaryTrajectories import SummaryTrajectories
import matplotlib.pyplot as plt
import pandas as pd

# Use the SummaryTrajectories class
fig_size = (12, 6)
summary_file = "RegionBornholm_01042019_31032020_FishCargTank_1_315569220_0_trajectories_summary.csv"
summary_trajectories = SummaryTrajectories(summary_file, save_figures=True, plot_figures=True,
                                           fig_size=fig_size, date="DateTimeStart")
df = summary_trajectories.df
df["Track length (sec)"] = pd.to_datetime(df["DateTimeEnd"]) - pd.to_datetime(df["DateTimeStart"])
df["Track length (sec)"] = df["Track length (sec)"].dt.total_seconds().astype(int)
df["Ship type"] = df["ShipType"] 
df.sample(n=2)
```

```python
# Basic data exploration
summary_trajectories.basic_statistics(resample_frequency=0)
```

```python
df.sort_values("Track length (sec)")
```

# Visualising distributions

```python
# Plot the trajectory count for each ship type
summary_trajectories.hist_bar_plot(df["ShipType"].value_counts().reset_index(name="counts"),
                                   "Bar", "counts", "index", file_name="trajectory_count_by_shipType",
                                   xlabel="Trajectory Count", ylabel="Ship type")

# Plot the number of unique MMSIs for each ship type
data = df.groupby("ShipType")["MMSI"].nunique().reset_index(name="counts").sort_values("counts", ascending=False)
summary_trajectories.hist_bar_plot(data, "Bar", "counts", "ShipType", file_name="unique_MMSIs_count_by_shipType",
                                   xlabel="Unique MMSIs", ylabel="Ship type")
```

```python
# Plot the mean speed histogram for each ship type
df_hist = df.reset_index()
hue_order = ["Fishing", "Cargo", "Tanker"]
summary_trajectories.hist_bar_plot(df_hist, "Histogram", "MeanSpeed", 
                                   file_name="mean_speed_histogram_by_shipType", xlabel="Mean speed (knots)",
                                   hue="Ship type", hue_order=hue_order, xlim=[-1, 22])

# Plot the mean speed histogram for cargo and tankers
summary_trajectories.hist_bar_plot(df_hist.loc[df_hist["ShipType"] != "Fishing"], 
                                   "Histogram", "MeanSpeed", xlabel="Mean speed",
                                   hue="ShipType", hue_order=["Cargo", "Tanker"], xlim=[-1, 22])
```

```python
# Plot the mean speed histogram for each ship type as three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))
df.loc[df["ShipType"] == "Fishing"]['MeanSpeed'].plot.hist(ax=ax1, bins=50) 
ax1.set_xlabel('Mean speed for fishing vessels')
ax1.set_ylabel('Number of Trajectories')
ax1.set_xlim(0, 11)
df.loc[df["ShipType"] == "Cargo"]['MeanSpeed'].plot.hist(ax=ax2, bins=50) 
ax2.set_xlabel('Mean speed for cargo vessels')
ax2.set_ylabel('Number of Trajectories')
ax2.set_xlim(0, 25)
df.loc[df["ShipType"] == "Tanker"]['MeanSpeed'].plot.hist(ax=ax3, bins=50) 
ax3.set_xlabel('Mean speed for tankers')
ax3.set_ylabel('Number of Trajectories')
ax3.set_xlim(0, 20)
plt.show()
```

```python
# Plot trajectory length counts for each ship type
summary_trajectories.hist_bar_plot(df_hist, "Histogram", "Track length (sec)", hue="Ship type", hue_order=hue_order)

# Plot trajectory length counts for each ship type - Zoomed in more
summary_trajectories.hist_bar_plot(df_hist, "Histogram", "Track length (sec)", 
                                   file_name="trajectory_length_counts_histogram_by_shipType", 
                                   hue="Ship type", hue_order=hue_order, xlim=[0, 100000])
```

```python
# Plot trajectory length counts for each ship type
summary_trajectories.hist_bar_plot(df_hist, "Histogram", "TrackLength",
                                   xlabel="Track length", hue="ShipType", hue_order=hue_order)

# Plot trajectory length counts for each ship type - Zoomed in more
summary_trajectories.hist_bar_plot(df_hist, "Histogram", "TrackLength", 
                                   file_name="trajectory_length_counts_histogram_by_shipType", 
                                   xlabel="Track length", hue="ShipType", hue_order=hue_order, xlim=[0, 6000])
```

```python
# Look at the mean track length by day of month
mean_track_length = df.groupby('Day of Month')['TrackLength'].mean().reset_index(name='Mean track length')
summary_trajectories.hist_bar_plot(mean_track_length, "Bar", x="Day of Month", y="Mean track length")

# Plot the median track length by day of month
median_track_length = df.groupby('Day of Month')['TrackLength'].median().reset_index(name='Median track length')
summary_trajectories.hist_bar_plot(median_track_length, "Bar", x="Day of Month", y="Median track length")
```

# Visualising time series data

```python
# Look at some basic time series analysis based on daily trajectory count grouped by ShipType
df_date_type = summary_trajectories.df_date_type
df_date_type.sample(n=3)
```

```python
# Plot the daily trajectory count for each ship type 
summary_trajectories.time_series_line_plot(df_date_type[["Cargo", "Tanker", "Fishing"]], 
                                           "Pandas", file_name="daily_trajectory_count_by_shipType",
                                           xlabel="Date", ylabel="Daily Trajectory Count")

# Plot the daily trajectory count for fishing vessels 
summary_trajectories.time_series_line_plot(df_date_type[["Fishing"]], "Pandas", 
                                           file_name="daily_trajectory_count_fishing",
                                           xlabel="Date", ylabel="Daily Trajectory Count")
```

```python
# Zoom in on June-September and show fishing vessels with the weekday name
df_zoom = df_date_type.loc[df_date_type["Month"].isin([6, 7, 8, 9])]
summary_trajectories.time_series_line_plot(df_zoom, "Seaborn", x=df_zoom.index, y="Fishing", 
                                           file_name="daily_trajectory_count_fishing_June_August",
                                           xlabel="Date", ylabel="Daily Trajectory Count", include_points=True)

# Zoom in on October-December and show fishing vessels with the weekday name
df_zoom = df_date_type.loc[df_date_type["Month"].isin([10, 11, 12])]
summary_trajectories.time_series_line_plot(df_zoom, "Seaborn", x=df_zoom.index, y="Fishing", 
                                           file_name="daily_trajectory_count_fishing_June_August",
                                           xlabel="Date", ylabel="Daily Trajectory Count", include_points=True)
```

```python
# Further explore the seasonality of the data with box plots to display the distributions for each group.
# First group the data by month to visualize yearly seasonality
summary_trajectories.fig_size = (16, 12)
summary_trajectories.box_plot(df_date_type, x="Month", ylabel="Trajectory Count",
                              file_name="daily_trajectory_count_by_shipType_month")

# Group the data by weekday to visualize weekly fluctuations
summary_trajectories.box_plot(df_date_type, x="Weekday Name", ylabel="Trajectory Count",
                              file_name="daily_trajectory_count_by_shipType_weekday")
```
